```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
ur1="https://drive.google.com/file/d/1dKZxapT3xLzOTOpy1LCyqctL8YEvzp4Y/view?usp=drive_link"
ur1='https://drive.google.com/uc?id=' + ur1.split('/')[-2]
df=pd.read_csv(ur1)
```

#### Task: Price Range vs. Online Delivery and Tabel Booking 

#### To analyzed this we need to plot 2 charts together 

```python
x1 = df['Has Online delivery']
y = df['Average Cost for two']
x2 = df['Has Table booking']
```

##### Plot 1 for Online Delivery with the avg cost
##### Plot 2 for Tabel Booking with the avg cost 

```python
plt.figure(figsize=(9,4))
plt.subplot(1, 2, 1)
sns.boxplot(x=x1, y=y)
plt.xlabel('Online Delivery')
plt.ylabel('Average Cost for Two')
plt.title('Impact of Online Delivery on Cost')
plt.subplot(1, 2, 2)
sns.boxplot(x=x2, y=y)
plt.xlabel('Table Booking')
plt.ylabel('Average Cost for Two')
plt.title('Impact of Table Booking on Cost')
plt.tight_layout()
plt.show()
```

```python
print("Both the online delivery and Table booking are negative impacted by the cost")
```

#### determine if higher-priced restaurants are more likely to offer these services 

```python
print("High priced restaurants which have average cost more than 10,000 did not provide Online delivery as well as Table Booking facilities")
```

##### lets check for the resturants have avg cost less than 10,000

```python
plt.figure(figsize=(9,4))
plt.subplot(1, 2, 1)
sns.boxplot(x=x1, y='Average Cost for two',data=df.loc[df['Average Cost for two']<10000])
plt.title('Online Delivery by cost')
plt.subplot(1, 2, 2)
sns.boxplot(x=x2, y= 'Average Cost for two',data=df.loc[df['Average Cost for two']<10000])
plt.title('Table booking by cost')
plt.tight_layout()
plt.show()
```

```python
print("The Online delivery facility is present for only those restaurant,where the avg cost is less than 1000")
print("The Table booking facility is available in restaurants with an average cost ranging form 1000 to 2000")
```
